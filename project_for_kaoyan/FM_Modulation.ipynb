```python
import matplotlib.pylab as pyl
```

```python
# time between samples
sampling_period = 0.0001

# x axis
ts = pyl.arange(0,1,sampling_period)

# modulation signal
ym = pyl.sin(2.0* pyl.pi*1.0*ts)
pyl.plot(ts,ym)
pyl.show()
```

```python
# FM modulation
# carrier properties
frenq = 100
modulation_factor = 15.0

# FM signal
yc = pyl.sin(2.0* pyl.pi* (frenq+modulation_factor*ym)*ts)
```

```python
# show FM signal
pyl.plot(ts,yc)
pyl.show()
```
